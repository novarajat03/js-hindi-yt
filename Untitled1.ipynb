```python
# Install pytesseract
!pip install pytesseract

# Install Tesseract OCR engine
!apt-get update
!apt-get install -y tesseract-ocr

# Optional: install additional language packs (e.g., Hindi, German, etc.)
# !apt-get install -y tesseract-ocr-hin tesseract-ocr-deu

# Check Tesseract version
!tesseract --version

import fitz  # PyMuPDF
import pdfplumber  # For text-based tables
import pandas as pd
import openpyxl
import cv2
import pytesseract
from PIL import Image
import numpy as np

def extract_tables_from_pdf(pdf_path, excel_output):
    tables_dict = {}
    table_count = 0

    # Step 1: Extract Tables Using pdfplumber
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            tables = page.extract_tables()

            for table in tables:
                if not table:
                    continue

                df = pd.DataFrame(table)  # Convert to DataFrame
                df = df.dropna(how='all')  # Remove empty rows
                table_name = f"Table_{table_count + 1}"
                tables_dict[table_name] = df
                table_count += 1

    # Step 2: If No Tables Found, Try OCR on Image-Based PDFs
    if not tables_dict:
        doc = fitz.open(pdf_path)
        for page_num in range(len(doc)):
            page = doc[page_num]
            pix = page.get_pixmap()
            img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

            # Convert Image to Text Using OCR
            img_np = np.array(img)
            gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
            text = pytesseract.image_to_string(gray)

            # Convert OCR text into table format (Basic Attempt)
            lines = text.split("\n")
            table_data = [line.split() for line in lines if line.strip()]
            if table_data:
                df = pd.DataFrame(table_data)
                table_name = f"OCR_Table_{table_count + 1}"
                tables_dict[table_name] = df
                table_count += 1

    # Step 3: Save to Excel
    if tables_dict:
        with pd.ExcelWriter(excel_output, engine='openpyxl') as writer:
            for sheet_name, table_df in tables_dict.items():
                table_df.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"✅ Extraction complete! Data saved to {excel_output}")
    else:
        print("❌ No tables found in the PDF!")

# Usage
pdf_path = "test3.pdf"  # Replace with your PDF file
excel_output = "extracted_tables.xlsx"
extract_tables_from_pdf(pdf_path, excel_output)
```
